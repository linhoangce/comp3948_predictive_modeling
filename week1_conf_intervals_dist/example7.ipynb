```python
import pandas as pd, numpy as np, time

N = 1_000_000
df = pd.DataFrame({
    "network": np.random.choice(["AT&T","Verizon","T-Mobile"], size=N),
    "index": np.arange(N),
    "duration": np.random.randint(1, 1000, size=N)
})

# Original approach
start = time.time()
count = df.groupby("network")["index"].count()
mean = df.groupby("network")["duration"].mean()
max_ = df.groupby("network")["duration"].max()
min_ = df.groupby("network")["duration"].min()
std = df.groupby("network")["duration"].std()
end = time.time()
print("Separate groupbys:", end - start)

# Aggregated approach
start = time.time()
stats = df.groupby("network").agg(
    **{
        "# Call": ("index", "count"),
        "Duration Mean": ("duration", "mean"),
        "Duration Max": ("duration", "max"),
        "Duration Min": ("duration", "min"),
        "Duration Std.": ("duration", "std"),
    }
)
end = time.time()
print("Single agg:", end - start)
```
